```python
def rule_(dfData):
    
    c1 = 8 ; c2 = -0.99 ;   c3 = 0.9973

    
    dcta = dfData[dfData['Label'] == 0].groupby(['Sender ID'])[dfGbs.amountUSDColLabel].agg('mean').to_dict()
    amount_means = dfData.apply(lambda x: dcta[x['Sender ID']] if x['Sender ID'] in dcta else 0 , axis=1)
    q1 = (dfData[dfGbs.amountUSDColLabel] - amount_means ) / amount_means ######### <<-- Relative Change of Sent Amount  : (amount - mean)/ mean
    
    
    dfData[dfGbs.timeDiffAssetColLabel] = dfGbs.dfGrpBySender[dfGbs.timeColLabel].transform('diff').dt.total_seconds()
    dcttd = dfData[dfData['Label'] == 0].groupby(['Sender ID'])[dfGbs.timeDiffAssetColLabel].agg('mean').to_dict()
    td_means = dfData.apply(lambda x: dcttd[x['Sender ID']] if x['Sender ID'] in dcttd else 0 , axis=1)
    q2 = (dfData[dfGbs.timeDiffAssetColLabel] - td_means ) / td_means ###### <<-- Relative Change of Time Difference between Transactions  : (time_diff - mean(time_diff)) / mean(time_diff)

    d_t1 = dfData[dfData['Label'] == 0].groupby(['Sender ID', 'Receiver ID'])[dfGbs.timeColLabel].agg('max').to_dict()
    d_t2 = dfData[dfData['Label'] == 0].groupby(['Sender ID', 'Receiver ID'])[dfGbs.timeColLabel].agg('min').to_dict()

    max_ = dfData.apply(lambda x: d_t1[(x['Sender ID'], x['Receiver ID'])] if (x['Sender ID'], x['Receiver ID']) in d_t1 else datetime.datetime.min , axis=1)
    min_ = dfData.apply(lambda x: d_t2[(x['Sender ID'], x['Receiver ID'])] if (x['Sender ID'], x['Receiver ID']) in d_t2 else datetime.datetime.min , axis=1)
    
    max_ = pd.to_datetime(max_, utc=True, errors = 'coerce')
    min_ = pd.to_datetime(min_ , utc=True, errors = 'coerce') 
   
    interv = (max_ - min_).dt.total_seconds()/(60*60*24) ###<<<----  we need to convert it to days instead of seconds!!!!!!

    count_tsx  = dfGbs.AggByReceiver(colName = dfGbs.amountUSDColLabel, grpLabel = None, subGrpLabel = None, calcType = CalcType.TYPE_COUNT)
    freq_of_tsx = interv / count_tsx
    dfData['Frequency of Transactions'] = freq_of_tsx
    freq_avg = dfData.groupby(['Sender ID'])['Frequency of Transactions'].transform('mean')
    q3 = (freq_of_tsx - freq_avg)/freq_avg
    
    r1 = (q1>c1) ;  r2 = (q2<c2) ;  r3 = (q3<c3)
    r = r1 & r2 & r3
    return r 
```

```python
hKFoldSplt = StratifiedGroupKFold(n_splits = numKFolds, shuffle = True, random_state = randomState)
for vTrainIdx, vTestIdx in hKFoldSplt.split(mX, vY, groups = dfData['Sender ID']):
    mXTrain, mXTest, vYTrain, vYTest = mX.iloc[vTrainIdx], mX.iloc[vTestIdx], vY.iloc[vTrainIdx], vY.iloc[vTestIdx]
    #xgbModel =XGBClassifier(tree_method="gpu_hist", max_depth = 20, feature_names = lTotalFeatures, feature_types = feature_types_, random_state=seedNum, enable_categorical=True) #XGBClassifier(use_label_encoder = False)
    xgbModel =XGBClassifier(n_estimators=250, tree_method="hist", max_depth = 20,  random_state=seedNum, enable_categorical=True)
    xgbModel.fit(mXTrain, vYTrain)
    vYPred = xgbModel.predict(mXTest)
    DisplayConfusionMatrix(vYTest, vYPred, lClasses = xgbModel.classes_)
    print(GenClassifierSummaryResults(vYTest, vYPred))
```
